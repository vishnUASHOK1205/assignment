```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 2: Load the dataset
df = pd.read_csv("House Price India.csv")

# Step 3: Perform Univariate Analysis
# Example: Histogram of house prices
sns.histplot(data=df, x="Price")
# Add more univariate visualizations as needed

# Step 3: Perform Bivariate Analysis
# Example: Scatter plot of living area versus house price
sns.scatterplot(data=df, x="living area", y="Price")
# Add more bivariate visualizations as needed

# Step 3: Perform Multivariate Analysis
# Example: Correlation matrix heatmap

corr_matrix = df.corr()
plt.figure(figsize=(15, 13))
sns.heatmap(corr_matrix, annot=True)
# Add more multivariate visualizations as needed

# Step 4: Perform Descriptive Statistics
# Example: Summary statistics of numerical columns
descriptive_stats = df.describe()
print(descriptive_stats)

# Step 5: Handle Missing Values
# Example: Check for missing values
missing_values = df.isnull().sum()
print(missing_values)

# Example: Fill missing values with mean
df.fillna(df.mean(), inplace=True)
# Add more handling of missing values as needed

# Save the updated dataset
df.to_csv("House Price India_updated.csv", index=False)
```
